```python
import pandas as pd
import matplotlib.pyplot as plt

# Data sizes and execution times
data_sizes = ['16M', '8M', '4M', '2M', '1M', '512K', '256K', '128K']
java_times = [1.8844, 0.8122, 0.3407, 0.1344, 0.0820, 0.0359, 0.0173, 0.0084]
tornado_times = [1.1482, 0.6072, 0.3798, 0.2488, 0.1774, 0.1419, 0.1333, 0.1255]
openmp_times = [0.2191, 0.0426, 0.0149, 0.0066, 0.0032, 0.0017, 0.0012, 0.0001]

# Creating a DataFrame
df = pd.DataFrame({
    'Data Size': data_sizes,
    'Java': java_times,
    'TornadoVM': tornado_times,
    'OpenMP': openmp_times
})

# Setting the 'Data Size' column as the index
df.set_index('Data Size', inplace=True)

# Plotting the data
plt.figure(figsize=(10, 6))
plt.plot(df['Java'], label='Java', marker='o')
plt.plot(df['TornadoVM'], label='TornadoVM', marker='s')
plt.plot(df['OpenMP'], label='OpenMP', marker='^')

# Adding labels and title
plt.xlabel('Data Size')
plt.ylabel('Execution Time (seconds)')
# plt.title('Execution Time for BFS Algorithm Across Different Data Sizes')
plt.legend()

# Inverting x-axis to show larger datasets on the left
plt.gca().invert_xaxis()

plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for RTX 4090
algorithms_4090 = ['bfs', 'lud', 'nw', 'particlefilter', 'srad', 'hotspot', 'pathfinder']
execution_time_4090_tornado = [1.1482, 10.5903, 8.5883, 0.9217, 31.0169, 58.5059, 0.7790]

# Data for QUADRO GP100
algorithms_gp100 = ['bfs', 'lud', 'nw', 'particlefilter', 'srad', 'hotspot', 'pathfinder']
execution_time_gp100_tornado = [3.5172, 57.3933, 38.6942, 2.234, 44.3431, 122.3213, 2.0133]

# Creating bar chart
fig, ax = plt.subplots(figsize=(14, 8))
width = 0.35  # width of the bars

indices = np.arange(len(algorithms_4090))  # the x locations for the groups

# Bars for RTX 4090
tornado_bars_4090 = ax.bar(indices - width/2, execution_time_4090_tornado, width, label='TornadoVM (RTX 4090)')

# Bars for QUADRO GP100
tornado_bars_gp100 = ax.bar(indices + width/2, execution_time_gp100_tornado, width, label='TornadoVM (QUADRO GP100)', color='orange')

# Annotate each bar with its y-value above the bar
for bar in tornado_bars_4090:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom')

for bar in tornado_bars_gp100:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, f'{yval:.2f}', ha='center', va='bottom')

ax.set_xlabel('Benchmark')
ax.set_ylabel('Execution Time (s)')
# ax.set_title('Execution Time Comparison Between RTX 4090 and QUADRO GP100')
ax.set_xticks(indices)
ax.set_xticklabels(algorithms_4090, rotation=45)
ax.legend()

plt.show()
```
